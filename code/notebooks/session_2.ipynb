```python
#Create a pandas dataframe with two columns, one named key and the
#other named data1 which contains a range of numbers.

df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'], 'data1':range(1,8,1)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a','b','d'], 'data2': range(3)})
df2
```

```python
pd.merge(df1, df2, how = "inner")
#all keys in both represented
```

```python
#Full
pd.merge(df1, df2, how = "outer")
#all keys in either
```

```python
pd.merge(df1, df2, how = "left")
#all keys in left reprsented
```

```python
pd.merge(df1, df2, how = "right")
#all keys in right represented
```
