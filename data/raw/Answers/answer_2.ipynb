```python
#which countries are available?
df1.loc[:,"Area"].unique() 
```

```python
df1 = df1.set_index("Area")
df1
```

```python
#Select the Netherlands
nld1 = df1.loc["Netherlands"]
nld1
```

```python
#select wheat
wheatNLD = nld1["Item"] == "Wheat"
wheatNLD
```

```python
nld2 = nld1[wheatNLD]
nld2
```

```python
#selct yield "Element"
wheatNLDYield = nld2["Element"] == "Yield"
wheatNLDYield
```

```python
nld3 = nld2[wheatNLDYield]
```

```python
#which columns do I have now
nld3.columns
```

```python
mylist = []
for c in nld3.columns:
    if 'F' not in c:
        mylist.append(c)
        
mylist
```

```python
#the following code uses list comprehension

#choose those column names that don't have an F in them
cols = [c for c in nld3.columns if 'F' not in c]
cols
```

```python
#choose those columns with a Y in them
cols = [c for c in cols if 'Y' in c]
cols
```

```python
nld3.head()
```

```python
#tranpose data (make columns rows)
nld4 = nld3[cols].T
nld4
```

```python
#replace the Y with nothing (so get rid of Y)
years = [c.replace('Y', '') for c in cols]

#then change string to int
years = [int(y) for y in years]
years
```

```python
#make a new column
nld4['years'] = years
nld4
```

```python
#set years as index
nld5 = nld4.set_index("years")
nld5
```

```python
#use matplotlib to plot
plt.figure(figsize=(20,10))
plt.plot(nld5.index, nld5['Netherlands'])
plt.title('NLD Wheat Yields')
plt.ylabel('hg/ha')
fig1 = plt.gcf()
plt.show()
fig1.savefig("NLD_Wheat_Yields.png")
```
